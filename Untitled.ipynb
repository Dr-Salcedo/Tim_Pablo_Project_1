```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
import pylab
import scipy.stats as stats
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

#Original Dataset
df = pd.read_csv('kc_house_data.csv')

#Dataset Changes
df = df.replace('?' , np.nan)
df['age'] = 2019-df['yr_built']

#Turning Price to Thousands of Dollars
df['price1000'] = df['price']/1000

#Dataset for < $1,000,000 Homes
df_Mill = df.loc[df['price'] < 1000000]
df_Mill.head()

dfX= df_Mill[['sqft_living', 'bathrooms', 'floors', 'sqft_lot', 'grade', 'sqft_living15', 'sqft_lot15', 'age', 'zipcode', 'view']]
dfY= df_Mill['price']
```

```python
dfXDummy1 = pd.get_dummies(dfX, columns=['zipcode', 'view'])
dfXDummy1.describe()
```

```python
dfXDummy2 = dfXDummy1.drop(['zipcode_98001', 'view_0.0'], axis=1)
```

```python
from sklearn.linear_model import Ridge
#Regression Model
X = dfXDummy2# here we have 2 variables for multiple regression. If you just want to use one variable for simple linear regression, then use X = df['Interest_Rate'] for example.Alternatively, you may add additional variables within the brackets
Y = dfY

X = sm.add_constant(X) # adding a constant
model = sm.OLS(Y, X).fit()
model.summary()
```
